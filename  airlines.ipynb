### DATA CLEANING

### importing the necessary libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### reading the data from the provided link

```python
A=pd.read_csv('archive/train.csv')
B=pd.read_csv('archive/test.csv')
A.head(10)
```

### checking the head of the data

```python
B.head()
```

```python
df = A.append(B, ignore_index=True)
df
```

### droping the unnecessary extra column 

```python
df=df.drop('Unnamed: 0',axis=1)
df.head()
```

### exploring the data

```python
df.describe()
```

```python
df.info()
```

### checking for null values

```python
df.isnull().sum()
```

### Filling the missing values with mean

```python
#Finding the mean of the column having NaN
mean_value=df['Arrival Delay in Minutes'].mean()
  
# Replace NaNs in column with the
# mean of values in the same column
df['Arrival Delay in Minutes'].fillna(value=mean_value, inplace=True)
```

### Converting the categorical variables 

```python
df['Gender']=pd.get_dummies(df['Gender'],drop_first=True)
df['Class']=pd.get_dummies(df['Class'],drop_first=False)
df['Customer Type']=pd.get_dummies(df['Customer Type'],drop_first=True)
df['Type of Travel']=pd.get_dummies(df['Type of Travel'],drop_first=True)
df['satisfaction']=pd.get_dummies(df['satisfaction'],drop_first=True)
```

### DATA VISUALIZATION

```python
sns.set_palette("GnBu_d")
sns.barplot(y='Arrival Delay in Minutes',x='satisfaction',data=df)
#by this we can observe that as Arrival Delay in Minutes increases, the dissatisfaction also increases
```

```python
sns.barplot(y='Cleanliness',x='satisfaction',data=df)
# by this we can observe that as cleanliness increases, satisfaction increases
```

```python
sns.boxplot(data=df,x='satisfaction',y='Inflight wifi service')
# by this we observe that as Inflight wifi service increases, satisfaction increases 
```

```python
sns.heatmap(df.corr())
```

## Dividing the data into attributes and labels

```python
X=df.drop('satisfaction',axis=1)
y=df['satisfaction']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

### FEATURE SCALING

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Random forest model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error


model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(X_train, y_train)
preds = model.predict(X_test)
print("Mean Absolute Error:",mean_absolute_error(y_test, preds))
```

### Accuracy

```python
from sklearn.metrics import r2_score
```

```python
print('r2_score :{}'.format(r2_score(y_test,preds)))
```

Hence, the acccuracy of the model is 88.5%
